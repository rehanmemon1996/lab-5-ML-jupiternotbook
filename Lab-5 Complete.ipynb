```python
from sklearn.datasets import load_iris
from pandas.plotting import parallel_coordinates

import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib
```

```python
plt.style.use('ggplot')
```

```python
data = load_iris()
```

```python
# load up SKLearn's iris dataset into a pandas dataFrame

df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
print(df.head(10))
```

```python
df['target_names'] = [data.target_names[i] for i in data.target]

# creates a series named as target name in df based on the 
# class of each observation is stored in the 
# .target attribute of the dataset.
```

```python
## Parallel Coordinates Start Here:

plt.figure()

parallel_coordinates(df,"target_names")

plt.show()
```

```python
from sklearn.datasets import load_iris
from pandas.plotting import andrews_curves

import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
data = load_iris()
print(data)
```

```python
data.feature_names
```

```python
data.target_names
```

```python
# now going to load up SKLearn's Iris Dataset into a Pandas dafaframe

df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df["target_names"] = [data.target_names[i] for i in data.target]
```

```python
## Andrews Curves Start Here:

plt.figure()
andrews_curves(df, 'target_names')

plt.show()
```
